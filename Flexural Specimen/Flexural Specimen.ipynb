```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data from CSV files
def load_csv(filename):
    data = pd.read_csv(filename)
    return data
```

```python
# Constants
L = 60  # Length
b = 22.6000  # Width
d = 12.3100  # Thickness

# Load CSV files
f_data = load_csv('K-P-K 2 F.csv')
s_data = load_csv('K-P-K 2 S.csv')

# Calculate stress and strain
f_data['Stress'] = 3 * f_data['0'] * L / (2 * b * d**2)
s_data['Strain'] = 6 * s_data['0'] * d / L**2

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(s_data['Strain'], f_data['Stress'],  linestyle='-')

plt.title('(K-P-K) Flexural Stress vs. Strain (Specimen 2)')
plt.xlabel('Flexural Strain')
plt.ylabel('Flexural Stress (σ)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV files
def load_csv(filename):
    data = pd.read_csv(filename)
    return data

# Constants
L = 60  # Length

# Data and labels for each specimen
specimens = [
    {'b': 18.7000, 'd': 11.8900, 'label': '(P-K-P) Specimen 3', 'f_file': 'P-K-P 3 F.csv', 's_file': 'P-K-P 3 S.csv'},
    {'b': 22.4500, 'd': 7.9500, 'label': '(P-K) Specimen 2', 'f_file': 'PK 2 F.csv', 's_file': 'PK 2 S.csv'},
    {'b': 22.6000, 'd': 12.3100, 'label': '(K-P-K) Specimen 2', 'f_file': 'K-P-K 2 F.csv', 's_file': 'K-P-K 2 S.csv'}
]

# Create a single plot
plt.figure(figsize=(8, 6))

for spec in specimens:
    # Load CSV files
    f_data = load_csv(spec['f_file'])
    s_data = load_csv(spec['s_file'])

    # Calculate stress and strain
    f_data['Stress'] = 3 * f_data['0'] * L / (2 * spec['b'] * spec['d']**2)
    s_data['Strain'] = 6 * s_data['0'] * spec['d'] / L**2

    # Plotting
    plt.plot(s_data['Strain'], f_data['Stress'], linestyle='-', label=spec['label'])

plt.title('Flexural Stress vs. Strain')
plt.xlabel('Flexural Strain')
plt.ylabel('Flexural Stress (σ)')
plt.grid(True)
plt.legend()
plt.show()
```
